# **Data Cleaning**
  

*   First, we find the values of nan, then we replace it with a suitable value.
*   In the next step, we check the data types and change them if needed.

*   In the last step, if there is a duplicate row, we delete it from the data.






```python
team_dataframe['residence'] = team_dataframe['residence'].fillna("blank")

team_dataframe['birthplace'] = team_dataframe['birthplace'].fillna("blank")

team_dataframe['height'] = team_dataframe['height'].fillna(team_dataframe['height'].mean())

team_dataframe['weight'] = team_dataframe['weight'].fillna(team_dataframe['weight'].mean())

team_dataframe['plays'] = team_dataframe['plays'].fillna(method='ffill')

team_dataframe['turned_pro'] = team_dataframe['turned_pro'].fillna("blank")

team_dataframe['current_prize'] = team_dataframe['current_prize'].fillna("blank")

team_dataframe['total_prize'] = team_dataframe['total_prize'].fillna("blank")

team_dataframe['current_rank'] = team_dataframe['current_rank'].fillna(team_dataframe['current_rank'].mean())
team_dataframe['current_rank'] = team_dataframe['current_rank'].astype(int)

team_dataframe['country'] = team_dataframe['country'].fillna("blank")
```

```python
team_score_dataframe['current_score'] = team_score_dataframe['current_score'].fillna(method='ffill')
team_score_dataframe['current_score'] = team_score_dataframe['current_score'].astype(int)

team_score_dataframe['display_score'] = team_score_dataframe['display_score'].fillna(method='ffill')
team_score_dataframe['display_score'] = team_score_dataframe['display_score'].astype(int)

team_score_dataframe['period_1'] = team_score_dataframe['period_1'].fillna(method='ffill')
team_score_dataframe['period_1'] = team_score_dataframe['period_1'].astype(int)

team_score_dataframe['period_2'] = team_score_dataframe['period_2'].fillna(method='ffill')
team_score_dataframe['period_2'] = team_score_dataframe['period_2'].astype(int)

team_score_dataframe['period_3'] = team_score_dataframe['period_3'].fillna(team_score_dataframe['period_3'].mean())
team_score_dataframe['period_3'] = team_score_dataframe['period_3'].astype(int)

team_score_dataframe['period_4'] = team_score_dataframe['period_4'].fillna(team_score_dataframe['period_3'].mean())
team_score_dataframe['period_4'] = team_score_dataframe['period_4'].astype(int)

team_score_dataframe['period_1_tie_break'] = team_score_dataframe['period_1_tie_break'].fillna(team_score_dataframe['period_1_tie_break'].mean())
team_score_dataframe['period_1_tie_break'] = team_score_dataframe['period_1_tie_break'].astype(int)

team_score_dataframe['period_2_tie_break'] = team_score_dataframe['period_2_tie_break'].fillna(team_score_dataframe['period_2_tie_break'].mean())
team_score_dataframe['period_2_tie_break'] = team_score_dataframe['period_2_tie_break'].astype(int)

team_score_dataframe['period_3_tie_break'] = team_score_dataframe['period_3_tie_break'].fillna(team_score_dataframe['period_3_tie_break'].mean())
team_score_dataframe['period_3_tie_break'] = team_score_dataframe['period_3_tie_break'].astype(int)
```

```python
tournament_dataframe['tournament_unique_id'] = tournament_dataframe['tournament_unique_id'].fillna("blank")
```

```python
tournament_dataframe['ground_type'] = tournament_dataframe['ground_type'].fillna("blank")
tournament_dataframe['tennis_points'] = tournament_dataframe['tennis_points'].fillna(tournament_dataframe['tennis_points'].mean())
tournament_dataframe['tennis_points'] = tournament_dataframe['tennis_points'].astype(int)
```

```python
venue_dataframe['country'] = venue_dataframe['country'].fillna("blank")
```

```python
odds_dataframe['winnig'] = odds_dataframe['winnig'].fillna(method='bfill')
odds_dataframe['winnig'] = odds_dataframe['winnig'].astype(bool)
```

```python
statistics_dataframe['home_total'] = statistics_dataframe['home_total'].fillna(statistics_dataframe['home_total'].mean())
statistics_dataframe['home_total'] = statistics_dataframe['home_total'].astype(int)

statistics_dataframe['away_total'] = statistics_dataframe['away_total'].fillna(statistics_dataframe['away_total'].mean())
statistics_dataframe['away_total'] = statistics_dataframe['away_total'].astype(int)

statistics_dataframe['statistic_type'] = statistics_dataframe['statistic_type'].astype(bool)

statistics_dataframe['value_type'] = statistics_dataframe['value_type'].astype(bool)
```

## What is the average duration of matches?


```python

match_time_info = time_dataframe

match_time_info.fillna(0, inplace=True)

match_time_info['total_duration'] = match_time_info[['period_1', 'period_2', 'period_3', 'period_4', 'period_5']].sum(axis=1)

valid_matches = match_time_info[match_time_info['total_duration'] > 0]

average_duration_seconds = valid_matches['total_duration'].mean()

average_duration_timedelta = pd.to_timedelta(average_duration_seconds, unit='s')
average_duration_formatted = str(average_duration_timedelta)

print("Average duration of matches:", average_duration_formatted)
```

##What is the average number of games per set in men's matches compared to women's matches?

```python
away_team = ht_dataframe
home_team = at_dataframe

teams = pd.concat([away_team, home_team], ignore_index=True)

teams['gender'] = teams['gender'].map({'M': 'male', 'F': 'female'})

teams = teams.drop_duplicates(subset=['match_id'])

df_power = power_dataframe

max_game_nums = df_power.groupby(['match_id', 'set_num'])['game_num'].max().reset_index()

max_game_nums['set_num'] = max_game_nums['set_num'].map({1: 'set_1', 2:'set_2', 3: 'set_3'})

max_game_nums_pivoted = max_game_nums.pivot(index='match_id', columns='set_num', values='game_num')

max_game_nums_pivoted.fillna(0, inplace=True)

teams_plus_infos = pd.merge(teams, max_game_nums_pivoted , on='match_id', how="inner")

grouped = teams_plus_infos.groupby('gender')

color_base = "#303030"

grouped = grouped[["set_1","set_2","set_3"]].mean()

# Plot the data
ax = grouped.plot(kind='bar')
ax.set_xlabel('Gender', color=color_base)
ax.set_ylabel('Number games',labelpad=10,color=color_base)
ax.set_xticklabels(grouped.index, rotation=0)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_color(color_base)
ax.spines['left'].set_color(color_base)

max_value = grouped.max().max()
ax.set_yticks(np.arange(0, max_value , 1))
plt.yticks(fontsize=10)

ax.yaxis.set_minor_locator(plt.MultipleLocator(5))

ax.set_title('', pad=20, color=color_base)

ax.set_axisbelow(True)
ax.grid(axis='y', color=color_base)

legend = ax.legend(loc='upper right',title="")
for text in legend.get_texts():
    text.set_color("#101010")

ax.set_ylim(top=ax.get_ylim()[1] + 2)


plt.show()
```

## Which player has the highest winning percentage against top 10 ranked opponents?

```python
win = odds_dataframe[odds_dataframe['winnig']==1]
merged_df = pd.merge(win, team_dataframe, on='match_id', how='left')
winner = merged_df.groupby('name').agg({'winnig' : 'size'})
top_10_winners = winner.nlargest(10, 'winnig')
```

```python
top_10_players = top_10_winners.index.tolist()
top_matches = merged_df[merged_df['name'].isin(top_10_players)]
player_matches = top_matches.groupby('name')['match_id'].nunique()
player_wins = top_matches[top_matches['winnig'] == 1].groupby('name')['winnig'].size()
player_win_percentage = (player_wins / player_matches).sort_values(ascending=False)
top_player = player_win_percentage.idxmax()

print("Player with the highest winning percentage against top 10 opponents:", top_player)
```

First, I get the names of the winning people by merging data frame Win and data frame team_dataframe, and then I find the 10 people who have the most wins with Nelargest. Finally, by comparing the winners against the top ten players, the person who won the most percentage is found.

## What is the average duration of matches per gender?

```python
match_time_info.fillna(0, inplace=True)

match_home_team_info = ht_dataframe
match_away_team_info = at_dataframe

home_away_merged = pd.merge(match_home_team_info, match_away_team_info, on='match_id', suffixes=('', '_away'))

home_away_merged.drop(columns=['gender_away'], inplace=True)

merged_data = pd.merge(match_time_info, home_away_merged, on='match_id', how='left')

merged_data['total_duration'] = merged_data[['period_1', 'period_2', 'period_3', 'period_4', 'period_5']].sum(axis=1)

valid_matches = merged_data[merged_data['total_duration'] > 0]

average_duration_per_gender = valid_matches.groupby('gender')['total_duration'].mean()/60

# Plotting
base_color = 'skyblue'

plt.figure(figsize=(8, 6))

average_duration_per_gender.plot(kind='bar', color=base_color)

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.xlabel('Gender')
plt.ylabel('Average Duration (minutes)')
plt.title('Average Duration of Matches per Gender', loc="left", pad=20)

plt.xticks(rotation=0)

plt.show()
```
